```python
import torch

# 행렬 출력해보기
A = torch.tensor([[1, 2],[3, 4], [5, 6]], dtype = torch.float32)

A_transpose = A.T

print("원래 행렬 A:")
print(A)
print(f"A의 크기: {A.shape}")

print("\n전치된 행렬 A^T:")
print(A_transpose)
print(f"A^T의 크기: {A_transpose.shape}")
```

```python
# 선형 변환 해보기
import torch

x = torch.ones(784, 1) # 784x1 열벡터. 모든 값 1

W = torch.randn(10, 784) # 10x784 가중치 행렬
b = torch.randn(10, 1) # 10x1 편향 벡터

y = torch.matmul(W, x) + b # y=Wx + b y: 10x1 행렬

print("--- 실험 결과 ---")
print(f"입력 벡터 크기: {x.shape}")
print(f"가중치 행렬 크기: {W.shape}")
print(f"출력 벡터 크기: {y.shape}")
print("\n출력 벡터(y)의 값:")
print(y)
```
